```python
#!/usr/bin/env python
import subprocess
import sys

if __name__ == "__main__":
    if len(sys.argv) < 2:
        print("Usage: python run.py [ingest|query] [options]")
        sys.exit(1)
    command = sys.argv[1]
    if command == "ingest":
        subprocess.run(["python", "src/ingest.py"] + sys.argv[2:])
    elif command == "query":
        subprocess.run(["python", "src/query.py"] + sys.argv[2:])
    else:
        print(f"Unknown command: {command}")
```
